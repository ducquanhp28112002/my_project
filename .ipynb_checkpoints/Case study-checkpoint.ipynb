```python
from tkinter.ttk import *
from tkinter import *
from tkinter import messagebox
from PIL import ImageTk, Image
import os
import time

root = Tk()
root.title("Restaurant Management System")
root.geometry("1600x780+0+0")

# Tạo thời gian thực
localtime = time.asctime(time.localtime(time.time()))

# Tạo 3 vùng trống
frame1 = Frame(root, width = 1600, height = 50)
frame1.pack()
frame2 = Frame(root, width = 900, height = 780)
frame2.pack(side = LEFT, fill = Y)
frame3 = Frame(root, width = 700, height = 780)
frame3.pack()

# Chèn ảnh minh họa
img = ImageTk.PhotoImage(Image.open("kfc.png"))
panel = Label(frame3, image = img)
panel.pack(side = "bottom", fill = "both", expand = "yes")

# Lời chào
label_welcome = Label(root, text = "Welcome to Queen Restaurant!", font = ("Arial", 30, "bold"), fg = "chocolate")
label_welcome.pack(side = TOP)

# Quảng cáo
label_bonus = Label(root, text = "If you buy more than 50 $, you will get 10% discount.", font = ("Arial", 20, "italic"), fg = "indianred")
label_bonus.pack()

# Hiển thị tên nhà hàng và thời gian thực
label_title = Label(frame1, font = ("Arial", 30, "bold"), text = "Queen Restaurant", fg = "red")
label_title.grid(column = 0, row = 0)
label_time = Label(frame1,font = ("Arial", 20, "italic"), text = localtime, fg = "steel blue")
label_time.grid(column = 0, row = 1)

# Đồ ăn
label_food = Label(frame2, text = "Food", font = ("Arial", 20, "bold"), bd = 10)
label_food.grid(column = 0, row = 0)

fried_chicken = StringVar()
fried_chicken.set("0")
french_fries = StringVar()
french_fries.set("0")
hamburgers = StringVar()
hamburgers.set("0")
cheese_sticks = StringVar()
cheese_sticks.set("0")
spaghetti = StringVar()
spaghetti.set("0")
cake = StringVar()
cake.set("0")
ice_cream = StringVar()
ice_cream.set("0")

label1 = Label(frame2, text = "Fried Chicken", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label1.grid(column = 0, row = 1)
entry1 = Entry(frame2, textvariable = fried_chicken, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry1.grid(column = 1, row = 1)

label2 = Label(frame2, text = "French Fries", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label2.grid(column = 0, row = 2)
entry2 = Entry(frame2, textvariable = french_fries, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry2.grid(column = 1, row = 2)

label3 = Label(frame2, text = "Hamburgers", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label3.grid(column = 0, row = 3)
entry3 = Entry(frame2, textvariable = hamburgers, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry3.grid(column = 1, row = 3)

label4 = Label(frame2, text = "Cheese Sticks", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label4.grid(column = 0, row = 4)
entry4 = Entry(frame2, textvariable = cheese_sticks, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry4.grid(column = 1, row = 4)

label5 = Label(frame2, text = "Spaghetti", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label5.grid(column = 0, row = 5)
entry5 = Entry(frame2, textvariable = spaghetti, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry5.grid(column = 1, row = 5)

label6 = Label(frame2, text = "Cake", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label6.grid(column = 0, row = 6)
entry6 = Entry(frame2, textvariable = cake, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry6.grid(column = 1, row = 6)

label7 = Label(frame2, text = "Ice Cream", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label7.grid(column = 0, row = 7)
entry7 = Entry(frame2, textvariable = ice_cream, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry7.grid(column = 1, row = 7)

# Đồ uống
label_drink = Label(frame2, text = "Drink", font = ("Arial", 20, "bold"), bd = 10)
label_drink.grid(column = 2, row = 0)

coca_cola = StringVar()
coca_cola.set("0")
pepsi = StringVar()
pepsi.set("0")
milo = StringVar()
milo.set("0")
sprite = StringVar()
sprite.set("0")
fanta = StringVar()
fanta.set("0")

label8 = Label(frame2, text = "Coca Cola", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label8.grid(column = 2, row = 1)
entry8 = Entry(frame2, textvariable = coca_cola, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry8.grid(column = 3, row = 1)

label9 = Label(frame2, text = "Pepsi", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label9.grid(column = 2, row = 2)
entry9 = Entry(frame2, textvariable = pepsi, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry9.grid(column = 3, row = 2)

label10 = Label(frame2, text = "Milo", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label10.grid(column = 2, row = 3)
entry10 = Entry(frame2, textvariable = milo, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry10.grid(column = 3, row = 3)

label11 = Label(frame2, text = "Sprite", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label11.grid(column = 2, row = 4)
entry11 = Entry(frame2, textvariable = sprite, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry11.grid(column = 3, row = 4)

label12 = Label(frame2, text = "Fanta", font = ("Arial", 16, "bold"), fg = "orange", bd = 10)
label12.grid(column = 2, row = 5)
entry12 = Entry(frame2, textvariable = fanta, font = ("Arial", 16, "bold"), bd = 10, bg = "salmon")
entry12.grid(column = 3, row = 5)

# Tạo menu (đơn giá)
def menu():
    window1 = Tk()
    window1.title("Menu")
    window1.geometry("400x280+900+500")
    
    # Đơn giá đồ ăn
    label_food = Label(window1, text = "Food", font = ("Arial", 20, "bold"), bd = 10)
    label_food.grid(column = 0, row = 0)
    
    label1 = Label(window1, text = "Fried Chicken:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label1.grid(column = 0, row = 1)
    
    label2 = Label(window1, text = "French Fries:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label2.grid(column = 0, row = 2)
    
    label3 = Label(window1, text = "Hamburgers:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label3.grid(column = 0, row = 3)
    
    label4 = Label(window1, text = "Cheese Sticks:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label4.grid(column = 0, row = 4)
    
    label5 = Label(window1, text = "Spaghetti:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label5.grid(column = 0, row = 5)
    
    label6 = Label(window1, text = "Cake:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label6.grid(column = 0, row = 6)
    
    label7 = Label(window1, text = "Ice Cream:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label7.grid(column = 0, row = 7)
    
    label8 = Label(window1, text = "10 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label8.grid(column = 1, row = 1)
    
    label9 = Label(window1, text = "3 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label9.grid(column = 1, row = 2)
    
    label10 = Label(window1, text = "5 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label10.grid(column = 1, row = 3)
    
    label11 = Label(window1, text = "5 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label11.grid(column = 1, row = 4)
    
    label12 = Label(window1, text = "8 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label12.grid(column = 1, row = 5)
    
    label13 = Label(window1, text = "4 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label13.grid(column = 1, row = 6)
    
    label14 = Label(window1, text = "3 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label14.grid(column = 1, row = 7)
    
    # Đơn giá đồ uống
    label_drink = Label(window1, text = "Drink", font = ("Arial", 20, "bold"), bd = 10)
    label_drink.grid(column = 3, row = 0)
    
    label15 = Label(window1, text = "Coca Cola:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label15.grid(column = 3, row = 1)
    
    label16 = Label(window1, text = "Pepsi:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label16.grid(column = 3, row = 2)
    
    label17 = Label(window1, text = "Milo:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label17.grid(column = 3, row = 3)
    
    label18 = Label(window1, text = "Sprite:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label18.grid(column = 3, row = 4)
    
    label19 = Label(window1, text = "Fanta:", font = ("Arial", 16, "bold"), fg = "dark orange")
    label19.grid(column = 3, row = 5)
    
    label20 = Label(window1, text = "2 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label20.grid(column = 4, row = 1)
    
    label21 = Label(window1, text = "2 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label21.grid(column = 4, row = 2)
    
    label22 = Label(window1, text = "1 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label22.grid(column = 4, row = 3)
    
    label23 = Label(window1, text = "2 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label23.grid(column = 4, row = 4)
    
    label24 = Label(window1, text = "2 $", font = ("Arial", 16, "bold"), fg = "dark orange")
    label24.grid(column = 4, row = 5)
    
    window1.mainloop()

button_menu = Button(frame2, bd = 10 , fg = "black", font = ("Arial" , 16, "bold"), width = 10, text = "PRICE", bg = "coral", command = menu)
button_menu.grid(column = 0, row = 8, pady = 10)

# Tạo order (đặt hàng)
def order():
    a = float(fried_chicken.get())
    b = float(french_fries.get())
    c = float(hamburgers.get())
    d = float(cheese_sticks.get())
    e = float(spaghetti.get())
    f = float(cake.get())
    g = float(ice_cream.get())
    h = float(coca_cola.get())
    i = float(pepsi.get())
    j = float(milo.get())
    k = float(sprite.get())
    l = float(fanta.get())

    cost_a = a * 10
    cost_b = b * 3
    cost_c = c * 5
    cost_d = d * 5
    cost_e = e * 8
    cost_f = f * 4
    cost_g = g * 3
    cost_h = h * 2
    cost_i = i * 2
    cost_j = j * 1
    cost_k = k * 2
    cost_l = l * 2
    
    # Tính tiền, kiểm tra có được giảm giá không và số tiền phải trả
    total = cost_a + cost_b + cost_c + cost_d + cost_e + cost_f + cost_g + cost_h + cost_i + cost_j + cost_k + cost_l
    total_bill.set(total)
    if total >= 50:
        sale_bill.set("10%")
        pay = total - (total * 0.1)
        pay_bill.set(pay)
    else:
        sale_bill.set("0")
        pay_bill.set(total)

button_order = Button(frame2, bd = 10 , fg = "black", font = ("Arial" , 16, "bold"), width = 10, text = "ORDER", bg = "coral", command = order)
button_order.grid(column = 1, row = 8, pady = 10)

# Xuất ra hóa đơn
def bill():
    window2 = Tk()
    window2.title("Bill")
    
    a = float(fried_chicken.get())
    b = float(french_fries.get())
    c = float(hamburgers.get())
    d = float(cheese_sticks.get())
    e = float(spaghetti.get())
    f = float(cake.get())
    g = float(ice_cream.get())
    h = float(coca_cola.get())
    i = float(pepsi.get())
    j = float(milo.get())
    k = float(sprite.get())
    l = float(fanta.get())

    cost_a = a * 10
    cost_b = b * 3
    cost_c = c * 5
    cost_d = d * 5
    cost_e = e * 8
    cost_f = f * 4
    cost_g = g * 3
    cost_h = h * 2
    cost_i = i * 2
    cost_j = j * 1
    cost_k = k * 2
    cost_l = l * 2
    
    total = cost_a + cost_b + cost_c + cost_d + cost_e + cost_f + cost_g + cost_h + cost_i + cost_j + cost_k + cost_l
    
    label_title = Label(window2, text = "Bill", font = ("Arial" , 20, "bold"), fg = "red")
    
    lb = Listbox(window2, height = 17, font = ("Arial" , 16, "bold"))
    
    if a != 0:
        lb.insert(1, f"Fried Chicken: {cost_a} $")
    if b != 0:
        lb.insert(2, f"French Fries: {cost_b} $")    
    if c != 0:
        lb.insert(3, f"Hamburgers: {cost_c} $")
    if d != 0:
        lb.insert(4, f"Cheese Sticks: {cost_d} $")
    if e != 0:
        lb.insert(5, f"Spaghetti: {cost_e} $")
    if f != 0:
        lb.insert(6, f"Cake: {cost_f} $")
    if g != 0:
        lb.insert(7, f"Ice Cream: {cost_g} $")
    if h != 0:
        lb.insert(8, f"Coca Cola: {cost_h} $")
    if i != 0:
        lb.insert(9, f"Pepsi: {cost_i} $")
    if j != 0:
        lb.insert(10, f"Milo: {cost_j} $")
    if k != 0:
        lb.insert(11, f"Sprite: {cost_k} $")
    if l != 0:
        lb.insert(12, f"Fanta: {cost_l} $")
        
    lb.insert(13, "----------------------------------")
    lb.insert(14, f"Total: {total} $")
    
    if total >= 50:
        lb.insert(15, "Sale: 10%")
        pay = total - (total * 0.1)
        lb.insert(16, f"You have to pay: {pay} $")
    else:
        lb.insert(17, f"You have to pay: {total} $")
        
    lb.insert(18, "THANK YOU!")
        
    label_title.pack()
    lb.pack()
    
    window2.mainloop()

button_bill = Button(frame2, bd = 10 , fg = "black", font = ("Arial" , 16, "bold"), width = 10, text = "BILL", bg = "coral", command = bill)
button_bill.grid(column = 3, row = 8, pady = 10)

# Reset lại toàn bộ sau khi hoàn thành
def reset():
    fried_chicken.set("")
    french_fries.set("")
    hamburgers.set("")
    cheese_sticks.set("")
    spaghetti.set("")
    cake.set("")
    ice_cream.set("")
    coca_cola.set("")
    pepsi.set("")
    milo.set("")
    sprite.set("")
    fanta.set("")
    total_bill.set("")
    sale_bill.set("")
    pay_bill.set("")

button_reset = Button(frame2, bd = 10 , fg = "black", font = ("Arial" , 16, "bold"), width = 10, text = "RESET", bg = "coral", command = reset)
button_reset.grid(column = 2, row = 8, pady = 10)

# Tạo exit khi thoát khỏi trang
def exit():
    a = messagebox.askyesno("Exit", "Do you want to exit?")
    if a == 1: 
        root.destroy()

button_exit = Button(root, bd = 10 , fg = "black", font = ("Arial" , 16, "bold"), width=10, text = "EXIT", bg = "coral", command = exit)
button_exit.pack(side = BOTTOM)

# Hiển thị số tiền, kiểm tra xem có được giảm giá không và số tiền phải trả
total_bill = StringVar()
sale_bill = StringVar()
pay_bill = StringVar()

label_total = Label(frame2, text = "Total", font = ("Arial", 20, "bold"), fg = "red", bd = 10)
label_total.grid(column = 0, row = 9, pady = 15)
entry_total = Entry(frame2, textvariable = total_bill, font = ("Arial", 16, "bold"), bd = 10, bg = "light coral")
entry_total.grid(column = 1, row = 9)

label_sale = Label(frame2, text = "Sale", font = ("Arial", 20, "bold"), fg = "red", bd = 10)
label_sale.grid(column = 0, row = 10, pady = 15)
entry_sale = Entry(frame2, textvariable = sale_bill, font = ("Arial", 16, "bold"), bd = 10, bg = "light coral")
entry_sale.grid(column = 1, row = 10)

label_pay = Label(frame2, text = "Pay", font = ("Arial", 20, "bold"), fg = "red", bd = 10)
label_pay.grid(column = 0, row = 11, pady = 15)
entry_pay = Entry(frame2, textvariable = pay_bill, font = ("Arial", 16, "bold"), bd = 10, bg = "light coral")
entry_pay.grid(column = 1, row = 11)

root.mainloop()
```
