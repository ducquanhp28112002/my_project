```python
import turtle as t

t.Screen()
t.bgcolor("cyan")
t.pensize(2)
t.penup()

# vẽ mặt trời
t.goto(170,280)
t.pendown()
t.forward(100)
t.backward(200)
t.goto(170,280)
t.left(90)
t.forward(100)
t.backward(200)
t.goto(170,280)
t.left(45)
t.forward(75)
t.backward(150)
t.goto(170,280)
t.right(90)
t.forward(75)
t.backward(150)
t.penup()
t.goto(205,245)
t.pendown()
t.fillcolor("yellow")
t.begin_fill()
t.circle(50)
t.end_fill()
t.penup()
t.goto(280,340)
t.write("Sun",font=("Arial","20"))

# vẽ cây
t.goto(250,20)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.right(90)
t.forward(100)
t.right(180-45)
t.forward(138)
t.right(180-45)
t.forward(100)
t.end_fill()
t.penup()
t.goto(250,-50)
t.pendown()
t.begin_fill()
t.right(90)
t.forward(100)
t.right(180-45)
t.forward(138)
t.right(180-45)
t.forward(100)
t.end_fill()
t.penup()
t.goto(250,-50)
t.penup()
t.goto(250,-120)
t.pendown()
t.begin_fill()
t.right(90)
t.forward(100)
t.right(180-45)
t.forward(138)
t.right(180-45)
t.forward(100)
t.end_fill()
t.penup()
t.goto(265,-190)
t.fillcolor("brown")
t.begin_fill()
t.pendown()
t.right(135)
t.forward(70)
t.right(90)
t.forward(30)
t.right(90)
t.forward(69)
t.right(90)
t.forward(30)
t.end_fill()
t.penup()
t.goto(300,-250)
t.write("Tree",font=("Arial","20"))

# vẽ nhà
t.goto(-300,-50)
t.fillcolor("blue")
t.pendown()
t.begin_fill()
t.forward(150)
t.right(90)
t.forward(170)
t.right(90)
t.forward(150)
t.right(90)
t.forward(170)
t.end_fill()
t.fillcolor("red")
t.begin_fill()
t.right(45)
t.forward(106)
t.right(90)
t.forward(106)
t.end_fill()
t.fillcolor("red")
t.begin_fill()
t.left(80)
t.forward(106)
t.left(100)
t.forward(106)
t.left(80)
t.forward(106)
t.left(100)
t.forward(106)
t.end_fill()
t.penup()
t.goto(-150,-220)
t.pendown()
t.fillcolor("yellow")
t.begin_fill()
t.left(80)
t.forward(106)
t.left(55)
t.forward(170)
t.left(125)
t.forward(106)
t.left(55)
t.forward(170)
t.end_fill()
t.penup()
t.left(125)
t.forward(30)
t.left(55)
t.forward(85)
t.pendown()
t.fillcolor("orange")
t.begin_fill()
t.right(55)
t.forward(40)
t.left(55)
t.forward(40)
t.left(125)
t.forward(40)
t.left(55)
t.forward(40)
t.end_fill()
t.penup()
t.goto(-250,-220)
t.right(180)
t.fillcolor("green")
t.begin_fill()
t.pendown()
t.forward(80)
t.right(90)
t.forward(40)
t.right(90)
t.forward(80)
t.end_fill()
t.penup()
t.goto(-300,70)
t.write("House",font=("Arial","20"))
t.goto(-230,-250)
t.right(180)
t.shape("turtle")
t.fillcolor("black")

t.done()
```
